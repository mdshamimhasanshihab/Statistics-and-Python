    calculate_cost_1_20, calculate_cost_21_50, and calculate_cost_51_100.
    2.Each function will take the number of guests as an argument.
    3.Calculate the total cost by multiplying the number of guests with the cost per guest.
    4.If the number of guests is above 20, and the customer is a VIP, apply a 5% discount on the total cost.

## 1c. Code

```python
# Include here the code for each problem. Remember to insert comments throughout! 
import random

# Define the three cost functions
def calculate_cost_1_20(guests):
    cost_per_guest = 70
    total_cost = guests * cost_per_guest
    return total_cost

def calculate_cost_21_50(guests):
    cost_per_guest = 60
    total_cost = guests * cost_per_guest
    return total_cost

def calculate_cost_51_100(guests):
    cost_per_guest = 50
    total_cost = guests * cost_per_guest
    return total_cost

# Define the main function
def calculate_total_cost():
    num_guests = random.randint(0, 100)
    print("Number of guests: ", num_guests)

    if num_guests <= 20:
        total_cost = calculate_cost_1_20(num_guests)
    elif num_guests > 20 and num_guests <= 50:
        total_cost = calculate_cost_21_50(num_guests)
    else:
        total_cost = calculate_cost_51_100(num_guests)

    is_vip = input("Are you a VIP client? (y/n): ")
    if is_vip == 'y' and num_guests > 20:
        total_cost *= 0.95
        print("5% discount applied for VIP client")

    print("Total cost: ", total_cost)

# Call the main function
calculate_total_cost()
```

## 2c. Code

```python
# Include here the code for each problem. Remember to insert comments throughout! 
# Create empty dictionary to store celebrity information
celebrities = {}
own_age = int(input("Enter Your age: "))
# Ask user to input name and age for each celebrity and store in dictionary
for i in range(3):
    name = input("Enter celebrity name: ")
    age = int(input("Enter celebrity age: "))
    celebrities[name] = age

# Calculate average age of celebrities
avg_age = sum(celebrities.values()) / 3

# Determine lowest and highest ages of celebrities
lowest_age = min(celebrities.values())
highest_age = max(celebrities.values())

# Calculate age differences and store in list
age_differences = []

for age in celebrities.values():
    age_differences.append(own_age-age)
print(f"Average age of celebrities: {avg_age}")   
# Loop through list of age differences and print message

for difference in age_differences:
    if difference > 0:
        print("I am old")
    else:
        print("I am young")
print(f"lowest age of celebrities: {lowest_age}")
print(f"Higest age of celebrities: {highest_age}")
print("Age difference according to celebrities: ")
print(age_differences)
```

## 3c. Code

```python
# Include here the code for each problem. Remember to insert comments throughout! 
Birthday_book = {'Sakib': 3, 'Dany': 5, 'Shami': 12, 'Nakib': 1, 'Roy': 7}

def increment_months(birthday_dict):
    new_dict = {}
    for key, value in birthday_dict.items():
        new_dict[key] = value + 1
    return new_dict

def extract_names(birthday_dict):
    name_list = []
    for key in birthday_dict.keys():
        name_list.append(key)
    name_list.sort()
    return name_list

def replace_months_with_AED(birthday_dict):
    new_dict = {}
    for key, value in birthday_dict.items():
        new_dict[key] = value * 100 # Replace the months with AED amounts
    return new_dict

print("Original Birthday Book: ", Birthday_book)

incremented_dict = increment_months(Birthday_book)
print("Birthday Book with incremented months: ", incremented_dict)

name_list = extract_names(Birthday_book)
print("Sorted List of Friends' Names: ", name_list)

AED_dict = replace_months_with_AED(Birthday_book)
print("Birthday Book with AED amounts: ", AED_dict)
```
